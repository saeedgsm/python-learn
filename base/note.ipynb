پایتون یک زبان برنامه نویسی محبوب است. این زبان توسط Guido van Rossum ساخته شد و در سال 1991 منتشر شد.
برای موارد زیر استفاده می شود:
•	توسعه وب (سمت سرور)،
•	توسعه نرم افزار,
•	ریاضیات
•	اسکریپت نویسی سیستم.
پایتون چه کاری می تواند انجام دهد؟
•	پایتون را می توان در سرور برای ایجاد برنامه های وب استفاده کرد.
•	پایتون را می توان در کنار نرم افزار برای ایجاد گردش کار استفاده کرد.
•	پایتون می تواند به سیستم های پایگاه داده متصل شود. همچنین می تواند فایل ها را بخواند و تغییر دهد.
•	پایتون را می توان برای مدیریت داده های بزرگ و انجام ریاضیات پیچیده استفاده کرد.
•	پایتون را می توان برای نمونه سازی سریع یا برای توسعه نرم افزار آماده تولید استفاده کرد.
